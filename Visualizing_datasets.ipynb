```python
import torch
import torch_geometric 
import torch_geometric.utils as ut
import networkx as nx
from matplotlib import pyplot as plt
edge_index = torch.tensor([[0, 1],
                           [1, 0],
                           [1, 2],
                           [2, 1],
                           [2, 0],
                           [0, 2],
                           [2, 3],
                           [3, 2],
                           [4, 3],
                           [3, 4]], dtype=torch.long)
x = torch.tensor([[12,5], [0,-2], [-5,3], [3,14], [2.5,0]], dtype=torch.float)

data = torch_geometric.data.Data(x=x, edge_index=edge_index.t().contiguous())
print(data)
print(data['x'])
print(data.has_isolated_nodes())
print(data.has_self_loops())
print(data.is_directed())
g = ut.to_networkx(data, to_undirected=True)
# nx.draw(g, with_labels= True)
# print("Planar")
# nx.draw_planar(g, with_labels = True)
print("Circular")
nx.draw_circular(g, with_labels = True)
# print("Random")
# nx.draw_random(g, with_labels = True)
# print("Spring")
# nx.draw_spring(g, with_labels = True)
# print("Spectral")
# nx.draw_spectral(g, with_labels = True)
```

```python
from torch_geometric.datasets import Planetoid

dataset = Planetoid(root='/tmp/Cora', name='Cora')
```

```python
data = dataset[0]
print(data.num_nodes)
print(data.num_edges)
print(data.num_features)
print(data.num_node_features)
print(data.num_edge_features)
print(data.has_isolated_nodes())
print(data.is_directed())
print(data.x.shape)
print(data.edge_index.shape)
# print(data.edge_attr)
print(data.y.shape)

colors = []
for i in range(data.num_nodes):
  if data.y[i]==0:
    colors.append("orange");
  elif data.y[i]==1:
    colors.append("red");
  elif data.y[i]==2:
    colors.append("green");
  elif data.y[i]==3:
    colors.append("violet");
  elif data.y[i]==4:
    colors.append("limegreen");
  elif data.y[i]==5:
    colors.append("blue");
  elif data.y[i]==6:
    colors.append("yellow");

g = ut.to_networkx(data, to_undirected=True)
nx.draw_networkx(g,node_color=colors, node_size=50, with_labels=False)

# nx.draw(g)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.manifold import TSNE

feature = np.random.randint(100, size=(100,1000))

tsne = TSNE(n_components=2, random_state=0, perplexity=10)
tsne_results = tsne.fit_transform(feature)

plt.figure(figsize=(6, 4))
plt.scatter(tsne_results[:,0], tsne_results[:,1], c=["red" if i<50 else "green" for i in range(100)])
```
